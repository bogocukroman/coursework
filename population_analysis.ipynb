# Прогнозирование численности населения по возрасту

Анализ датасета WPP Population

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('WPP_Population.csv')
df.head()
```

```
df = df[['Location', 'Time', 'AgeGrp', 'PopMale', 'PopFemale']]
df['PopTotal'] = df['PopMale'] + df['PopFemale']
df.head()
```

```
df['GrowthRate'] = df.groupby(['Location', 'AgeGrp'])['PopTotal'].pct_change() * 100
df['GrowthRate'] = df['GrowthRate'].fillna(df['GrowthRate'].mean())
df = df.drop_duplicates()
```

```
df['PopTotal'].hist()
plt.title('Распределение населения')
plt.show()
```

```
sns.scatterplot(data=df, x='Time', y='PopTotal')
plt.show()
```

```
corr = df[['PopMale','PopFemale','PopTotal','GrowthRate']].corr()
sns.heatmap(corr, annot=True)
plt.show()
```
